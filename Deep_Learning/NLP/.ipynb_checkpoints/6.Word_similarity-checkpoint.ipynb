# 단어 간의 유사도! ( 벡터 유사도 구하기 )

- red : L1 distance ( Manhattan distance )

# 1. L1 distance
- 맨허튼 distance
- 두 vector간의차원별 값의 차이의 절대값을 모두 합한 값

```python
def get_L1(x1,x2):
    return ((x1-x2).abs()).sum()
```

# 2. L2 distance
- 유클리디안 distance
- 최단 거리

```python
def get_L2(x1,x2):
    return ((x1-x2)**2).sum()**0.5
```

# 3. Infinity Norm
- 차원 별 값의 차이 중 가장 큰 값

```python
def get_Infdist(x1,x2):
    return ((x1-x2).abs()).max()
```

# 4. Cosine Similarity

- 두 벡터의 방향 & 크기 모두 고려!
- 두 벡터 사이의 요소 별 (element wise)곱을 사용 ( = 벡터의 내적 )
- 1에 가까울 수록 방향 일치
- 0에 가까울 수록 직교
- -1에 가까울 수록 방향 반대

### 희소 벡터(sparse vector)일 경우 큰 문제 발생! 직교하면서 곱의 값이 0이 되므로, 제대로 반영 X

```python
def get_cos_sim(x1,x2):
    return (x1*x2).sum() / ((x1**2).sum()**0.5 * (x2**2).sum()**0.5)
```

# 5. Jacard Similarity
- 집합 간의 유사도!

### 

```python
import torch
```

```python
def get_JC_sim(x1,x2):
    return torch.stack([x1,x2]).min(dim=0)[0].sum() / torch.stack([x1,x2]).max(dim=0).sum()
```
