# ELMo(Embeddings from Language Model)

### 사전 훈련된 언어모델을 사용함!

## 1. biLM(Bidirectional Language Model)의 사전 훈련 

- ElMO의 biLM은 "다층 구조(Multi-Layer)"형태를 띔 ( 최소 2개 이상의 hidden layer )
- embedding : 임베딩 방법으로 charCNN을 사용함 ( embedding이 글자(character) 단위로 이루어짐 )
- 유의점 : RNN에서 다루는 양방향 RNN과 ELMo에서의 biLM은 다름! 양방향 RNN은 순방향 RNN의 hidden state와 역방향 RNN의 hidden state을 연결한 이후 다음 층의 입력으로 보냄. BUT biLM은 "각각"의 hidden state만을 다음 hidden layer로 보내며, 훈련 시킨 후에 은닉 상태를 연결
