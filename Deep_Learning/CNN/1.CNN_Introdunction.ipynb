# 1. Introduction

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Conv2D, MaxPool2D,Dropout
```

```python
print(tf.__version__)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
#from tensorflow.keras.datasets import cifar10
from keras.datasets import cifar10
```

```python
(X_train,y_train),(X_test,y_test) = cifar10.load_data()
```

```python
class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']
```

```python
X_train.max(), X_train.min()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train.shape, X_test.shape
```

```python
plt.imshow(X_test[0])
```

```python
y_test
```

# 2. CNN Model Structure

### 1) Sequential

```python
model = Sequential()
```

### 2) Conv2D

```python
model.add(Conv2D(filters=32,kernel_size=(3,3),padding='same',
                 activation='relu',input_shape=[32,32,3]))
```

```python
model.add(Conv2D(filters=32,kernel_size=(3,3),padding='same',
                 activation='relu'))
```

### 3) MaxPool2D

```python
model.add(MaxPool2D(pool_size=(2,2), strides=2, padding='valid'))
```

### 4) Dropout

```python
model.add(Dropout(0.5)) # 0.5 만큼 "버리기"
```

### 5) Flatten

```python
model.add(Flatten())
```

### 6) Dense

```python
model.add(Dense(units=128,activation='relu'))
```

```python
model.add(Dense(units=10, activation='softmax'))
```

### [SUMMARY]

```python
model.summary()
```

# 3. Compile & Fit

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
history = model.fit(X_train[0:10000],y_train[0:10000],batch_size=32,epochs=5,verbose=1,
          validation_data=(X_test,y_test))
```

```python
model.evaluate(X_test,y_test)
```

# 4. Plot Accuracy & Loss

```python
epoch_range = range(1,6)
```

### 1) Accuracy Plot

```python
plt.plot(epoch_range, history.history['accuracy'])
plt.plot(epoch_range, history.history['val_accuracy'])
plt.title('Model Accuarcy ( Train & Validation )')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train','Valid'], loc='upper left')
plt.show()
```

### 2) Loss Plot

```python
plt.plot(epoch_range, history.history['loss'])
plt.plot(epoch_range, history.history['val_loss'])
plt.title('Model Loss ( Train & Validation )')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train','Valid'], loc='upper left')
plt.show()
```
