```python
def read_image(path):
    gfile = tf.io.read_file(path)
    image = tf.io.decode_image(gfile, dtype=tf.float32)
    return image
```

```python
plt.imshow(read_image(data_paths[3]))
plt.show()
```

```python
for i in range(4):
    plt.imshow(read_image(data_paths[i]))
    plt.show()
```

## Images in List

- 하나의 batch에 8개의 사진을 담을 경우
- (batch_size, height, width, channel )

```python
batch_images = []
```

```python
for path in data_paths[:8]:
    image = read_image(path)
    batch_images.append(image)
```

```python
plt.imshow(batch_images[1])
plt.show()
```

```python
batch = tf.convert_to_tensor(batch_images)
batch.shape # 8개의 32*32*3 사진
```

```python
def make_batch(batch_paths):
    batch_images = []
    
    for path in batch_paths:
        image = read_image(path)
        batch_images.append(image)
        
    return tf.convert_to_tensor(batch_images)
```

```python
batch_size  = 16
for step in range(4):
    batch_images = make_batch(data_paths[step*batch_size : (step+1)*batch_size])
    
    plt.imshow(batch_images[0]) # 각 배치의 1번째 image
    plt.show()
```
