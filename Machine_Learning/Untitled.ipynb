## Feature 해석
- SIDO_NM : 도시명 (unique - 서울)
- SIGUNGU_NM : '구' 명
- BJDONG_NM : '동' 명
- LAND_CD : 지역 고유 코드 ( SGG_CD(구) + BJDONG_CD(동) + LAND_GBN(토지형태) + BOBN(본번) + BUBN(부번) )
- LAND_GBN_NM : 토지 형태
- YEAR : 연도 ( unique - 2019 )
- BASE_MON : 기준 시점 (unique - 20190101 )
- JIGA : 지가
