```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.regplot('total_bill', 'tip', data=tips)
```

```python
mean = [2,55]
cov = [[1.1,0.4],[2.2,3]]
x_value, y_value = np.random.multivariate_normal(mean,cov, 100).T
```

```python
sns.regplot(x_value, y_value, color='purple')
```

```python
# 축 제목 지정하는 법
seriesx_value = pd.Series(x_value, name='Var-x')
seriesy_value = pd.Series(y_value, name='Var-y')
```

```python
sns.regplot(seriesx_value, seriesy_value, marker='D', color='gray')
```

```python
sns.regplot('total_bill','tip', data=tips, marker='D',
           scatter_kws={'marker':'D', 'color':'blue'}, 
            line_kws={'color':'red', 'linewidth':4.5})
```

```python
sns.regplot(x_value, y_value, ci=34)
```

```python
sns.regplot(x_value, y_value, ci=95)
```

```python
sns.regplot('size','total_bill', data=tips, x_jitter=0.1,
           scatter_kws={'color':'green'})
```

```python
sns.regplot('size','total_bill', data=tips, x_estimator=np.mean)
```
