```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month','year','passengers')
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False)
```

```python
sns.clustermap(flights, cmap='Blues_r', linewidth=1.5, figsize=(8,6))
```

```python
# standardize across col(1) or rows(0)
sns.clustermap(flights, standard_scale=1)
```

```python
# normalize our dataset = z_score / 0(rows) 1(cols)
sns.clustermap(flights, z_score=1)
```

```python
sns.clustermap(flights, z_score=0)
```
