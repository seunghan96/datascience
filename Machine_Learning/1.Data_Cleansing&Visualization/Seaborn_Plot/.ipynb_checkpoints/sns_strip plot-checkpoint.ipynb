```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips =sns.load_dataset('tips')
```

```python
# horizontal strip plot
sns.stripplot(tips['tip'], color='green')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True) ( # True = 1 )
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=0.2)
```

```python
sns.stripplot(x='total_bill', y='day', data=tips, jitter=0.3, linewidth=1.5)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, hue='sex', 
              palette='coolwarm', jitter=True, linewidth=2, split=True)
```

```python
sns.stripplot('day','total_bill','smoker',data=tips,
              jitter=True,split=True, palette='winter_r')
```

```python
sns.stripplot(x='sex', y='total_bill',hue='day', data=tips, 
              order=['Female','Male'], marker='D', size=10, 
              edgecolor='gray',alpha=0.4,palette='husl',
              jitter=True, split=True)
```

```python
# boxplot + strip plot
sns.stripplot('tip','day',data=tips,jitter=True, palette='winter_r', color='0.1')
sns.boxplot('tip','day',data=tips, palette='husl')
```

```python
sns.stripplot(x='tip',y='day',data=tips,jitter=True)
sns.violinplot(x='tip',y='day',data=tips, color='0.6')
```
