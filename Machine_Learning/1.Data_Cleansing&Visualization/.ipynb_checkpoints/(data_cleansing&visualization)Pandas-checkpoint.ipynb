### (1) Method Chaining

```python
# 예시 1
class foo():
    def __init__(self, kind=None):
        self.kind = kind
    def my_print(self):
        print (self.kind)
        return self
    def line(self):
        self.kind = 'line'
        return self
    def bar(self):
        self.kind='bar'
        return self
```

```python
a = foo()
a.line()
a.my_print()
a.bar()
a.my_print()
```

```python
a = foo().line().my_print().bar().my_print()
```

### (3) map VS apply VS applymap
#### 내가 만든 커스텀 함수(custom function)를 적용시키고 싶을 때! 주로 lamda와 함께 사용
#### 장점 : for문보다 빨라!

차이점?
- map() : (1차원)series 형태에서만! (+ element 중심 접근) 
- apply() : (1차원)series & (2차원) dataframe 모두 가능 (+ row.column 중심 접근)
- applymap() : (2차원)dataframe에서만! (+ element 중심 접근)

## 1. map

```python
def func(x):
    return round(x,-4)

data['population_total2']  = data['population_total'].map(lambda x : func(x))
```

```python
data['population_total2']  = data['population_total'].map(lambda x : round(x,-4))
```

```python
data = data.drop(['population_total2'],axis=1)
```
