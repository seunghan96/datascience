## 1. Import Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans ,DBSCAN , AgglomerativeClustering
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv').iloc[:,1:]
```

## 2. EDA

```python
iris['Species'].unique()
```

```python
sns.pairplot(iris)
```

```python
col = pd.Series(iris['Species']).map({'setosa':'r','versicolor':'b','virginica':'g'})
plt.scatter(iris['Petal.Width'],iris['Petal.Length'],c=col)
```

## 3.Clustering
### (1) KMeans

```python
kmeans = KMeans(n_clusters=3)
```

```python
# label(y)  
x=iris.iloc[:,0:4]
```

```python
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
col = pd.Series(kmeans.labels_).map({0:'r',1:'b',2:'g'})
```

```python
col.values
```

```python
plt.scatter(iris['Petal.Width'],iris['Petal.Length'],c=col)
```

### (2) DBSCAN

```python
dbscan = DBSCAN(eps=0.5,min_samples=5)
```

```python
dbscan.fit(x).labels_
```

```python
col = pd.Series(dbscan.fit(x).labels_).map({-1:'k',0:'r',1:'b'})
```

```python
plt.scatter(iris['Petal.Width'],iris['Petal.Length'],c=col)
```

### (3) Agglomerative

```python
agnes = AgglomerativeClustering(n_clusters=3)
```
