## 1. Import Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv('iris.csv').iloc[:,1:]
```

```python
iris['Species'].unique()
```

```python
plt.scatter(iris['Petal.Width'],iris['Petal.Length'])
```

```python
col = pd.Series(iris['Species']).map({'setosa':'r','versicolor':'b','virginica':'g'})
plt.scatter(iris['Petal.Width'],iris['Petal.Length'],c=col)
```

## 2. Clustering with K-means

```python
kmeans = KMeans(n_clusters=3)

col = pd.Series(kmeans.fit(iris.iloc[:,0:4]).labels_).map({0:'r',1:'b',2:'g'})
plt.scatter(iris['Petal.Width'],iris['Petal.Length'],c=col)
```

### elbow method to find appropriate 'k'

```python
sse = [] # sum of squares error

for i in range(1,10) :
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(iris.iloc[:,0:4])
    sse.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),sse)
plt.plot(np.linspace(3,3,6),np.linspace(-1,700,6),'--')
```
