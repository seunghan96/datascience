```python
card = card.dropna()
```

```python
X = card.iloc[:,1:30]
Y = card.iloc[:,30]
```

# Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,Y_train,Y_val = train_test_split(pd.DataFrame(X_scaled),Y,test_size=0.33,
                                              random_state=123,stratify=Y)
```

# Linear method
- 주성분 분석 ( PCA, Principal Component Analysis )
- 특잇값 분해 ( SVD, Singular Value Decomposition )
- 랜덤 투영 ( Random Projection )

```python
from sklearn.decomposition import PCA,IncrementalPCA,SparsePCA,KernelPCA
from sklearn.decomposition import TruncatedSVD
from sklearn.random_projection import GaussianRandomProjection, SparseRandomProjection
```

```python
def reduced_df(X_train,X_val,method,dim):
    X_train_reduced = pd.DataFrame(method.fit_transform(X_train),
                              index=X_train.index)
    X_val_reduced = pd.DataFrame(method.fit_transform(X_val),
                             index=X_val.index)
    
    X_train_reduced = pd.concat([X_train_reduced,Y_train],axis=1)
    X_val_reduced = pd.concat([X_val_reduced,Y_val],axis=1)
    colnames = ["PC" + str(i) for i in range(1,dim+1)]
    colnames.append('class')
    X_train_reduced.columns = colnames
    X_val_reduced.columns = colnames
    return X_train_reduced,X_val_reduced
```

## PCA (Principal Component Analysis)
( PCA, 점진적 PCA, 희소 PCA, 커널 PCA )

```python
def pca_evr_plot(pca_method):
    evr = pca_method.explained_variance_ratio_
    plt.figure(figsize=(50,20))
    plt.subplot(1, 2, 1)
    plt.bar(range(len(evr)),evr,alpha=0.9,align='center')
    plt.xlabel('Index of PC',fontsize=50)
    plt.ylabel('Explained Variance Ratio',fontsize=50)
    plt.title('Explained Variance Ratio of each PC',fontsize=60)
    
    plt.subplot(1, 2, 2)
    plt.bar(range(len(evr)),evr.cumsum(),alpha=0.9,align='center')
    plt.title('Cumulative Explained Variance Ratio',fontsize=60)
    plt.xlabel('Used number of PCs',fontsize=50)
    plt.ylabel('Cumulative Explained Variance Ratio',fontsize=50)
    plt.show()
```

### 일반 PCA

```python
n_components = X_train.shape[1]
random_state = 123

pca = PCA(n_components=n_components,random_state=random_state)
```

```python
pca_train, pca_val= reduced_df(X_train,X_val,pca,n_components)
```

```python
pca_evr_plot(pca)
```

### 점진적 PCA

```python
n_components = X_train.shape[1]
batch_size = None

inc_pca = IncrementalPCA(n_components=n_components,batch_size=batch_size)
```

```python
inc_pca_train, inc_pca_val = reduced_df(X_train,X_val,inc_pca,n_components)
```

```python
pca_evr_plot(inc_pca)
```

### 희소 PCA

```python
n_components = 6
random_state = 123
alpha = 0.0001
n_jobs = -1

sps_pca = SparsePCA(n_components=n_components,alpha=alpha,
                    random_state=random_state,n_jobs=n_jobs)
```

```python
sps_pca_train, sps_pca_val = reduced_df(X_train,X_val,sps_pca,n_components)
```

### 커널 PCA

```python
n_components = 6
random_state = 123
gamma = None
kernel = 'rbf'
n_jobs = 1

ker_pca = KernelPCA(n_components=n_components,kernel=kernel,gamma=gamma,
                   n_jobs=n_jobs,random_state=random_state)
```

```python
ker_pca_train, ker_pca_val = reduced_df(X_train,X_val,ker_pca,n_components)
```

```python
ev_ker = np.var(ker_pca_train,axis=0) 
evr_ker = ev_ker / np.sum(ev_ker)
pd.DataFrame(evr_ker).plot.bar()
```

## SVD (Singular Value Decomposition)

```python
from sklearn.decomposition import TruncatedSVD
```

```python
n_components = 6
algorithm = 'randomized'
n_iter = 10
random_state = 123
```

```python
svd = TruncatedSVD(n_components=n_components,algorithm=algorithm,
                   n_iter=n_iter,random_state=random_state)
```

```python
svd_train,svd_val = reduced_df(X_train,X_val,svd,n_components)
```

## Random Projection

```python
from sklearn.random_projection import GaussianRandomProjection, SparseRandomProjection
```

### GRP (Gaussian Random Projection)

```python
n_components = 5
random_state = 123
#eps = 0.99
```

```python
grp = GaussianRandomProjection(n_components=n_components,random_state=random_state)
```

```python
grp_train,grp_val = reduced_df(X_train,X_val,grp,n_components)
```

### SRP (Sparse Random Projection)

```python
n_components = 5
density = 'auto'
dense_output = False
random_state = 123
#eps = 0.99
```

```python
srp = SparseRandomProjection(n_components=n_components,density=density,
                             dense_output=dense_output,random_state=random_state)
```

```python
srp_train,srp_val = reduced_df(X_train,X_val,srp,n_components)
```
