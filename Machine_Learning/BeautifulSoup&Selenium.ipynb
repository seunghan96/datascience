### get

```python
# ? 앞 부분까지
main_url = 'https://search.naver.com/search.naver?'

param = {
    'where':'post',
    'sm':'tab_jum',
    'query':'%EC%8B%A0%EC%84%B8%EA%B2%BD'
}
```

```python
res5 = requests.get(url=main_url, params=param)
```

```python
res5
```
