# 실습 1. breast cancer
>### kernel : 'linear'

```python
from sklearn import datasets
cancer = datasets.load_breast_cancer()
```

```python
cancer.feature_names
```

```python
cancer.target_names
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target,test_size=0.3, random_state=42 )
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train,y_train)
```

```python
clf2 = SVC(kernel='rbf')
clf2.fit(X_train,y_train)
```

```python
clf3 = SVC(kernel='poly', degree=2)
clf3.fit(X_train,y_train)
```

```python
y_pred1 = clf.predict(X_test)
y_pred2 = clf2.predict(X_test)
y_pred3 = clf3.predict(X_test)
```

```python
print('Kernel : LINEAR')
print('training dataset accuracy :',clf.score(X_train,y_train).round(3))
print('test dataset accuracy :',clf.score(X_test,y_test).round(3),'\n')

print('Kernel : RBF')
print('training dataset accuracy :',clf2.score(X_train,y_train).round(3))
print('test dataset accuracy :',clf2.score(X_test,y_test).round(3),'\n')

print('Kernel : POLYNOMIAL with degree 2')
print('training dataset accuracy :',clf3.score(X_train,y_train).round(3))
print('test dataset accuracy :',clf3.score(X_test,y_test).round(3),'\n')
'''
print('Kernel : POLYNOMIAL with degree 4')
print('training dataset accuracy :',clf4.score(X_train,y_train).round(3))
print('test dataset accuracy :',clf4.score(X_test,y_test).round(3),'\n')
'''
```

```python
from sklearn.metrics import classification_report

print('Linear :', classification_report(y_test,y_pred1),'\n')
print('RBF :', classification_report(y_test,y_pred2),'\n')
print('POLY 2 :', classification_report(y_test,y_pred3))
```

```python
data.head()
```

```python
for col in cat_features:
    data = pd.concat([data, pd.get_dummies(data[col], prefix=col, prefix_sep=':')], axis=1)
    data.drop(col, axis=1, inplace=True)
    
data.head()
```

```python
X = data.iloc[:,data.columns!='y']
y = data['y']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
print(X_train.shape,X_test.shape )
```

```python
from sklearn.metrics import accuracy_score
```

```python
clf = SVC(kernel='rbf', gamma = 0.00001, C = 10)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_train)

accuracy_score(y_train,y_pred)
```

```python
y_pred2=clf.predict(X_test)
accuracy_score(y_test,y_pred2)
```

## 실습 3. SVM + Bagging
- SVM의 단점 : too long time! 
- Bagging의 base estimator로 SVM을 사용해보기

```python
from sklearn.svm import LinearSVC
from sklearn.ensemble import BaggingClassifier
```

```python
estimator = LinearSVC()

n_estimators = 10
n_jobs = 2 # 2 이상 : multi-processing ( 병렬 처리 )

model = BaggingClassifier(base_estimator=estimator,
                          n_estimators=n_estimators,
                          max_samples=1./n_estimators,
                          n_jobs=n_jobs)

model.fit(X_train,y_train)
```

```python
model.predict_proba(X_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('training dataset accuracy :',model.score(X_train,y_train).round(3))
print('test dataset accuracy :',model.score(X_test,y_test).round(3),'\n')
```
