```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
num = np.random.randn(150)  # N(0,1)따르는 150개의 난수
```

```python
sns.distplot(num)
```

```python
sns.distplot(num,color='red')
```

```python
# x축에 축이름 붙이기
label_dist = pd.Series(num,name='variable x')
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist, color='green', hist=False)
```

```python
sns.distplot(label_dist, color='green', hist=False, rug=True)
```
