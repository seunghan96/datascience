```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(3)
```

```python
# horizontal boxplot
sns.boxplot(tips['size'])
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
sns.boxplot('day','total_bill','time', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,order=['Sun','Sat','Fri','Thur'], palette='husl')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.boxplot(data=iris, palette='coolwarm')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, palette='husl')
sns.swarmplot(x='day',y='total_bill',data=tips, color='0.5')
```
