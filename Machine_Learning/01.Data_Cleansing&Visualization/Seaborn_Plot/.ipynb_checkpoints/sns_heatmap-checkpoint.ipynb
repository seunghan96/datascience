```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 0~1사이 값 랜덤하게 12*15 뽑기
normal = np.random.rand(12,15)
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal, annot=True)
```

```python
sns.heatmap(normal, vmin=-2, vmax=2)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights = flights.pivot('month','year','passengers')
flights
```

```python
sns.heatmap(flights, annot=True)
```

```python
sns.heatmap(flights,annot=True, fmt='d')
```

```python
sns.heatmap(flights, linewidths=0.9, cmap='summer', 
            annot=True, fmt='d', vmin=0,vmax=1000, center=flights.loc['June',1954])
```
