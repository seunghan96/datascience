```python
# PDF (Probability Distribution Function)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mean = [0,0]
cov =[[0.2,0], [0,3]]
```

```python
x_axis, y_axis = np.random.multivariate_normal(mean,cov,size=40).T
```

```python
sns.kdeplot(x_axis, shade=True,color='black')
```

```python
sns.kdeplot(x_axis, y_axis)
```

```python
sns.kdeplot(x_axis, y_axis, n_levels=18, cmap='RdBu')
```

```python
sns.kdeplot(x_axis, y_axis, n_levels=40, cmap='RdBu')
```

```python
sns.kdeplot(y_axis, bw=1)
```

```python
sns.kdeplot(y_axis, bw=0.2)
```

```python
sns.kdeplot(y_axis, vertical=True)
```

```python
iris = sns.load_dataset('iris')
```

```python
setosa = iris[iris.species == 'setosa']
versicolor = iris[iris.species == 'versicolor']
```

```python
sns.kdeplot(setosa.petal_length, setosa.petal_width, cmap='coolwarm', n_levels=15)
sns.kdeplot(versicolor.petal_length, versicolor.petal_width, cmap='RdBu', shade=True)
```
