```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.violinplot(tips['tip'])
```

```python
sns.violinplot(tips['total_bill'])
```

```python
# x,y 2 variables
sns.violinplot(x='size', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',hue='sex', data=tips)
```

```python
sns.violinplot('day','total_bill','smoker',data=tips,palette='husl')
```

```python
sns.violinplot('day','total_bill','smoker',data=tips,palette='husl', split=True)
```

```python
sns.violinplot(x='day',y='total_bill', hue='sex', data=tips, split=True,
               palette='coolwarm', order=['Sat','Thur','Sun','Fri'])
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',
              palette='RdBu', inner='quartile',split=True, scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',
              palette='RdBu', inner='quartile',split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',
              palette='RdBu', inner='quartile',scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker',
              palette='husl', inner='stick',split=True, scale='count', bw=0.1)
```
