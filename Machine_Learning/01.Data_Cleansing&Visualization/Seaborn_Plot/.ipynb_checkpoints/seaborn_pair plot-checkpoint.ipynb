```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(tips)
```

```python
# hue
sns.pairplot(iris, hue='species', palette='husl')
```

```python
# markers
sns.pairplot(iris, hue='species', markers=['o','D','s'])
```

```python
# vars
sns.pairplot(iris, vars=['sepal_length', 'sepal_width'])
```

```python
sns.pairplot(iris, size=4, x_vars=['sepal_length','sepal_width'], y_vars=['petal_length','petal_width'])
```

```python
sns.pairplot(iris, diag_kind='kde')
```

```python
sns.pairplot(iris, diag_kind='kde', hue='species', palette='husl', kind = 'reg')
```
