```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.hist, 'total_bill', color='green', bins=20)
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, row='smoker', col='time')
x = x.map(sns.regplot, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker')
x = x.map(plt.scatter, 'total_bill', 'tip')
```

```python
x = sns.FacetGrid(tips, col='time', hue='smoker')
x.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
x = sns.FacetGrid(tips, col='day', size=10, aspect=0.5)# aspect . 
x.map(sns.boxplot, 'time', 'total_bill', color='green')
```

```python
x = sns.FacetGrid(tips, col='day',hue='smoker',palette='husl', size=8, aspect=0.7)
x.map(plt.scatter, 'total_bill','tip').add_legend()
```
