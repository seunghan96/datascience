# Association Rule 

문제점: 제품의 개수가 k 개이면, 이런 조합이 2^k개! 너무 많아 -> 탐색 범위를 좁혀야! HOW?

사용 알고리즘 : Apriori Algorithm

대표적인 package : mlxtend, apyori..

## 1. Mlxtend의 apriori

### (1) Import Dataset
( 소비자들의 물품 구매 목록 / 물품은 숫자로 표시 )

```python
import pandas as pd
```

```python
items = pd.read_csv('items.csv',sep='delimiter', header=None)
items.head(5)
```

```python
items.shape
```

### (2) Data Preprocessing

목적 : mlxtend의 apriori에 데이터를 넣으려면, dataframe 형태로 변환해줘야! ( row : 고객, column : 물품, 값 : T/F )

(1) 위의 df형태를 우선 list형태로

(2) mlxtend.preprocessing의 TransactionEncoder를 통해 T/F로 변환

```python
itemlist = items.values.tolist()
itemlist
```

```python
itemlist2 = []
for i in range(len(itemlist)):
    itemlist2.append(itemlist[i][0].split(','))
    
itemlist2
```

### (3) Association anlaysis with (Mlxtend) apriori

( mlxtend의 transaction encoder를 통해, 특정 물품이 있으면 True / 없으면 False로 바꾸기 -> for 효율성 )

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(itemlist2).transform(itemlist2)
```

```python
te_ary
```

```python
te.fit(itemlist2)
```

```python
itemdf = pd.DataFrame(te_ary, columns=te.columns_)
itemdf
```

### (4) Rules
- 4. support값이 최소 0.1이상이 되는 물품(혹은 물품조합)만 찾아내기

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
items_10 = apriori(itemdf, min_support=0.1,use_colnames=True)
items_10
```

```python
items_10['length'] = items_10['itemsets'].apply(lambda x: len(x))
items_10
```

```python
association_rules(items_10, metric="confidence", min_threshold=0.7)
```

```python
association_rules(items_10, metric="lift", min_threshold=1.4)
```

## 2. Apyori의 apriori

### (1) Import Dataset
( 7501명의 20가지 물품에 대한 장바구니 data )

```python
data = pd.read_csv('store_data.csv',header=None)
data.head(10)
```

```python
data.shape
```

### (2) Data Preprocessing
- List 형태로 바꿔주기

```python
itemlist = []
for i in range(len(data)):  
    itemlist.append([str(data.values[i,j]) for j in range(len(data.columns))])
```

```python
itemlist
```

### (3) Association anlaysis with (Apyroi) apriori

```python
from apyori import apriori  
```

```python
association_rules = apriori(itemlist, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)  
association_results = list(association_rules)  

association_results
```

```python
association_results[0]
```

```python
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
