```python
clustering_dt.head()
```

- blog/community/news/twitter : 방송콘텐츠 관련 언급량

## 3. Clustering

```python
X = clustering_dt.loc[:,'10대':'man']
X.head()
```

```python
kmeans = KMeans()
```

```python
sse = []
for i in range(1,10) :
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
```

```python
sse
```

```python
plt.plot(range(1,10),sse)
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
clustering_dt.columns[1:]
```

```python
# 각 cluster의 중심점을 dataframe으로!
pd.DataFrame(kmeans.cluster_centers_, columns=clustering_dt.columns[1:], index=['1번','2번','3번','4번'])
```

```python
plt.scatter(clustering_dt['twitter'], clustering_dt['news'], c=kmeans.labels_, label='True Position')
```

```python
plt.scatter(clustering_dt['30대'], clustering_dt['SMR'], c=kmeans.labels_, label='True Position')
```

```python
Label = pd.DataFrame(kmeans.labels_,columns=['label'])
r = pd.concat([clustering_dt,Label],axis=1)
```

```python
r.head()
```

## 4. Result

```python
names=clustering_dt['program']
result=pd.concat([names,Label],axis=1)

result.head()
```

```python
label_unique = result['label'].unique()
grouping = result.groupby(['label'])

res = [ {'programs': result.loc[grouping.groups[l]]['program'].tolist()} for l in label_unique]
result_grouped = pd.DataFrame(res, index=[str(l) + '번' for l in label_unique])
result_grouped
```

```python
def doAgglomerative(X, nclust=4):
    model = AgglomerativeClustering(n_clusters=nclust, affinity = 'euclidean', linkage = 'ward')
    clust_labels1 = model.fit_predict(X)
    return (clust_labels1)

clust_labels1 = doAgglomerative(X, 4)
agglomerative = pd.DataFrame(clust_labels1)
X.insert((X.shape[1]),'agglomerative',agglomerative)

# 시각화
fig = plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter(X['10대'],X['search'],
                     c=agglomerative[0],s=50)

ax.set_title('Agglomerative Clustering')
ax.set_xlabel('10대')
ax.set_ylabel('search')
plt.colorbar(scatter)
```
