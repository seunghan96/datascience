# Prim's Algorithm

node A 선택 -> node A와 연결된 edge list 중, 최소 weight의 edge 연결된 node B선택 

-> node B에 연결된 edge list들을 아까의 edge list에 append -> 반복...

( 유의점 : cycle 안생기게끔 )

### kruskal과의 비교

- (both) greedy algorithm
- (Kruskal) weight가 가장 작은 edge부터 선택
- (Prim) 특정 node를 먼저 선택 후, 해당 node와 연결된 edge중 가장 weight가 가장 작은 것 선택
