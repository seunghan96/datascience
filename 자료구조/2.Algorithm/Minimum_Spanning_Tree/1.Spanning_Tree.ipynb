# Spanning Tree (신장 트리)

## 1. 신장 트리
- 그래프의 "모든 node"가 연결 + "트리의 속성"도 만족

- 신장 트리의 조건 : 본래의 그래프의 모든 노드를 포함 & 모든 노드가 서로 연결 & 트리의 속성 만족(cycle X)
