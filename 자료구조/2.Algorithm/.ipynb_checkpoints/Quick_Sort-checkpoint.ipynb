# Quick Sort

- Sorting의 핵심
- 기준점(pivot) 정하고, 기준점 보다 작으면 left, 크면 right
- 위 과정을 재귀적으로 반복

```python
def qsort(data):
    if len(data) <= 1:
        return data
    
    left,right = list(), list()
    pivot= data[0]
    
    for i in range(2,len(data)):
        if pivot > data[i]:
            left.append(data[i])
        else :
            right.append(data[i])
    
    total = qsort(left) + [pivot] + qsort(right)
    return total
```

```python
qsort([3,5,2,1,7,10,26,5,4])
```

```python
def qsort2(data):
    if len(data) <=1:
        return data
    
    pivot = data[0]
    left = [item for item in data[1:] if pivot > item]
    right = [item for item in data[1:] if pivot <= item]
    
    total = qsort(left) + [pivot] + qsort(right)
    return total
```

```python
qsort([3,5,2,1,7,10,26,5,4])
```

### 알고리즘 분석
- 병합 정렬과 유사, 시간 복잡도는 O(n*logn)
- worst case : O(n^2)
