# Dynamic Programming (동적 프로그래밍) & Divide and Conquer (분할 정복)

## 1. Dynamic Programming (DP)
- 상향식 접근법 (작은 문제 해결 -> 부분 문제의 해로 활용 -> 전체 문제 해결)
- Memorization : 이전에 계산한 값 저장 ( 다시 계산하지 않기 위해! 실행속도 Fast )
- ex) 피보나치 수열

## 2. Divide and Conquer 
- 문제를 나눌 수 없을 때 까지 나누어서 풀고, 다시 합병
- 하향식 접근법 
- 일반적으로 재귀함수 사용
- ex) 병합 정렬, 퀵 정렬

### 공통점 : 문제를 나눠서 해결함
### 차이점 : 상향식 vs 하향식
- DP : 부분 문제는 중복 됨 -> 상위 문제 해결 시 재활용
- D&C : 부분 문제 중복 X, Memorization X

### ex) 피보나치 수열

### 

1) Dynamic Programming
- 불필요한 반복 X

```python
def fibo_dp(n):
    li = [0 for i in range(n+1)]
    li[0] = 0
    li[1] = 1
    for index in range(2,n+1):
        li[index] = li[index-1] + li[index-2]
    return li[n]
```

```python
fibo_dp(10)
```

2) Divide and Conquer 
- Recursive Function 사용 
- 문제점 : 동일한 과정 굳이 반복?

```python
def fibo_dc(n):
    if n <= 1:
        return n
    return fibo_dc(n-1) + fibo_dc(n-2)
```

```python
fibo_dc(10)
```
