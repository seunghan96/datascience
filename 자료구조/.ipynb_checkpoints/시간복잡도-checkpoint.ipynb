# 알고리즘 복잡도

- 1) 시간 복잡도
- 2) 공간 복잡도

### 반복문이 시간 복잡도에 큰 영향!

## 알고리즘 성능 표기법

### 1) Big O 표기법 : O(N)
- 알고리즘 최악의 실행 시간 ( 최소한의 성능을 보장한다는 의미 )
- 가장 일반적으로 사용

### 2) 오메가 & 세타 표기법
- 오메가 : 최상
- 세타 : 평균

## [ Big O 표기법 ]

### Example
1부터 n까지의 합을 구하는 알고리즘

[방법 1]

```python
def summation(n):
    total = 0
    for i in range(1,n+1):
        total += i
    return total
```

```python
summation(10)
```

[방법 2] 

```python
def summation2(n):
    return int(n*(n+1)/2)
```

```python
summation2(10)
```

시간복잡도로 check! 

n의 변화에 따라 얼마나 변화??
- summation : O(n)
- summation2 : O(1)
