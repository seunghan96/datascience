```python
n,k = map(int,input().split()) # n : 물건 개수, k : 최대 무게
dp = [[0]*(k+1) for _ in range(n+1)]

for i in range(1,n+1): # (1) i번째 물건이
    w,v = map(int,input().split())  
    for j in range(1,k+1): # (2) 최대 무게 상한이 j일 때
        if j<w: # (상한을 넘겨버리면)
            dp[i][j] = dp[i-1][j] # (이제 물건을 더 추가 못하는 경우)
        else:
            cand1 = dp[i-1][j]
            cand2 = dp[i-1][j-w]+v
            dp[i][j] = max(cand1,cand2)
            
print(dp[n][k])
```
