## TIP
- DP를 위해선 "점화식" 세우기

D[i] = 길이가 i일 때의 경우의 수

길이가 i인 수열을 만드는 방법 
- (i-1) 길이 + 1
- (i-2) 길이 + 2

### D[i] = D[i-1] + D[i-2]

```python
n = int(input())

dp = [n]*1000001
dp[1] = 1
dp[2] = 2

for i in range(3,n+1):
    dp[i] = (dp[i-2] + dp[i-1]) % 15746 # 엄청 커지기 위해 방지
    
print(dp[n])
```
