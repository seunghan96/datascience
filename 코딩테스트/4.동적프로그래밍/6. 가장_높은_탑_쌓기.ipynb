LIS : Longest Increasing Sequence

```python
n = int(input())
array = [(0,0,0,0)]

for i in range(1,n+1):
    a,h,w = map(int, input().split())
    array.append((i,a,h,w))
    
array.sort(key=lambda x:x[3]) # weight 기준으로 정렬 ( 뒤에선 '무게' 고려 않고, '너비'만 고려할 수 있음 )

dp = [0]*(n+1)

for i in range(1,n+1): # 뒤 
    for j in range(0,i): # 앞
        if array[i][1] > array[j][1]:
            cand1 = dp[i] # 현재의 높이
            cand2 = dp[j]+array[i][2] # 이전 것의 높이 + 새로운 벽돌의 높이
            dp[i] = max(cand1, cand2) 

# 역추적
max_val = max(dp)
index = n
result = []

while index!=0:
    if max_val ==dp[index]:
        result.append(array[index][0]) # 몇 번째 block인지
        max_val -= array[index][2] # 높이
    index -=1
    
result.reverse()
print(len(result))
[print(i) for i in result]
```
