- 난이도 : 하
- 유형 : BFS & DFS

```python
n = int(input())
pair_num = int(input())
adj = [[] for _ in range(n+1)]
visited = [0]*(n+1)
count = 0

for _ in range(pair_num):
    x,y = map(int,input().split())
    adj[x].append(y)
    adj[y].append(x)
    
def dfs(now_pos):
    global count
    count +=1
    visited[now_pos] = 1
    for next_pos in adj[now_pos]:
        if not visited[next_pos]:
            dfs(next_pos)
            
dfs(1)
print(count-1) # 숙주 컴퓨터 제외!
```
