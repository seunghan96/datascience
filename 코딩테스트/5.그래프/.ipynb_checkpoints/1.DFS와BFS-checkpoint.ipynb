- 유형 : DFS, BFS

```python
from collections import deque
```

```python
def dfs(v):
    print(v, end=' ')
    visited[v] = 1
    for e in adj[v]: # Depth First
        if not visited[e]:
            dfs(e)
```

```python
def bfs(v):
    q = deque([v]) # deque : 양쪽에서 다 넣고 뺄 수 있음
    while q:
        v = q.popleft()
        if not visited[v]:
            visited[v] = 1
            print(v, end=' ')
            for e in adj[v]:
                if not visited[e]:
                    q.append(e)
```

```python
n,m,v = map(int,input().split()) # node 수, edge 수, 시작 정점 번호
adj = [ [] for _ in range(n+1)] # node간의 adjacency matrix

for _ in range(m): # edge 개수 전부
    x,y = map(int, input().split())
    adj[x].append(y)
    adj[y].append(x)
    
for e in adj:
    e.sort()
    
visited = [0]*(n+1)
dfs(v)
print()

visited = [0]*(n+1)
bfs(v)
```
