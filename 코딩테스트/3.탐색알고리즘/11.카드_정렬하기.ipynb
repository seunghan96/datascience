- 유형 : 힙, 자료 구조, 그리디

```python
import heapq
```

```python
n = int(input())
count = 0
card_deck = []

for _ in range(n):
    heapq.heappush(card_deck,int(input()))
    
while len(card_deck) !=1:
    one = heapq.heappop(card_deck)
    two = heapq.heappop(card_deck)
    sum_val = one + two
    count += sum_val
    heapq.heappush(card_deck,sum_val)
    
print(count)
```
