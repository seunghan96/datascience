위상 정렬 : 순서가 정해져 있는 작업에서, 순서를 정해주는 알고리즘

```python
import heapq
```

```python
n,m = map(int, input().split()) # n : 문제 수, m : 조건 수
array = [ []for i in range(n+1)]
indegree = [0] *(n+1)

heap = [] 
result = []

# 조건(array) & 진입 차수(indegree) 설정
for _ in range(m):
    first,second = map(int,input().split())
    array[first].append(second)
    indegree[second] +=1

for i in range(n+1):
    if indegree[i]==0: # 현재 제일 먼저 풀어야 할 (indegree가 0인) 문제들만 일단 heap에 담기
        heapq.heappush(heap,i) # 
        
while heap:
    data = heapq.heappop(heap) # 가장 작은 수 뽑기
    result.append(data)
    for y in array[data]:
        indegree[y] -=1
        if indegree[y] ==0:
            heapq.heappush(heap,y) # 한번 iteration이 끝나면, indegree가 1이었던 애들이 새롭게 indegree 0으로 되어서...
            
for i in result:
    print(i, end=' ')
```
