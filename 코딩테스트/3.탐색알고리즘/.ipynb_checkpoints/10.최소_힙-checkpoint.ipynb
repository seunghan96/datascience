- 난이도 : 하
- 유형 : 힙, 자료 구조

```python
import heapq
```

```python
n = int(input())
heap = []
result = []

for _ in range(n):
    num = int(input())
    if num==0:
        if heap:
            result.append(heapq.heappop(heap)) # (이미 있으면)가장 작은 수 출력 & result에 담는다
        else:
            result.append(0) # (비어있으면) result에 담는다
    else :
        heapq.heappush(heap,num)
        
for i in result:
    print(i)
```
