# [ Kruskal's Algorithm ]

- MST(Minimum Spanning Tree)의 방법 중 하나

### 순서

1) 모든 node을 독립적인 집합으로 만듬

2) 모든 edge의 weight을 기준으로 (작은순으로) 정렬 & weight가 작은 edge부터 비교

3) 두 node의 최상위 node을 확인하고, 서로 다를 경우 두 edge 연결 ( = cycle 생기지 않게! )

### Union-Find Algorithm 

기존에 n개의 node로 형성된 집합이 있음. 

이 상황에서 새로운 edge 추가시, 그 edge가 기존 집합 내의 2개의 node를 연결시킨다면 그것은 cycle!

### 순서
- 1) n개의 원소가 개별 집합으로 초기화
- 2) Union : 두 집합을 하나로 합침 ( 2 tree를 1 tree로 )
- 3) Find : 여러 node가 존재 시, 2개의 node 선택 후 그 두 node의 root node 확인! 같은 부분집합에 속하는지 check! ( cycle 판단 )

### union-by-rank 기법 ( union 시 )

- 각 tree의 rank를 기억
- union 시, 두 tree의 rank가 다르면, 작은 높이의 tree를 큰 높이의 tree에 붙임

높이가 같다면, 하나의 tree의 높이를 1 증가시켜주고, 그 곳에 작은 높이의 tree를 연결!

( 초기화 시, 모든 원소의 rank가 0 인 개별 집합 상태에서 )

- 높이가 h인 tree가 만들어지려면, 높이가 h-1인 tree 2개 필요
- 높이가 h-1인 tree 위해 n개 원소 필요하다면, 높이가 h인 tree 위해서는 최소 2n개 필요!

시간 복잡도를 O(logN)으로 줄일 수 있음!!
