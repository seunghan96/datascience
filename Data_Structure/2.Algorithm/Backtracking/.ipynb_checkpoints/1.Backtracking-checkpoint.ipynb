# Backtracking

제약 조건 만족 문제에서 해를 찾는 기법
- 제약 조건 충족 X 시, 즉시 backtrack ( 다시는 확인하지 않음 )

모든 경우의 수를 "상태 공간 트리(State Space Tree)"로 표현함

ex) 위에서 b가 조건을 충족 X다면, 굳이 밑까지 내려가지 않음!


상태 공간 트리를 탐색하면서, 제약이 맞지 않으면 다음 곳으로 넘어감
- Promising : 해당 루트가 조건에 맞는지 check
- Pruning : 조건에 맞지 않으면 포기하고 다른 루트로 돌아감 ( 시간 절약 O )

< 요약 : 트리 구조를 기반으로, DFS로 깊이 탐색 & Promising & Pruning >
