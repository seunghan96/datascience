## 4. Positional Encoding

RNN이 자연어 처리에서 많이 사용된 이유는, 단어를 "순차적으로 입력"받아 처리하는 RNN의 특징 때문! 이것이 각 단어의 위치 정보를 알려줬기 때문!

하지만, Transformer은 RNN을 사용하지 않는다. 대신에, Positional Encoding이라는 방법을 통해 단어의 위치(순서)를 파악한다. 단어의 위치 정보를 얻기 위해, 각 단어를 임베딩 한 이후, 이에 위치 정보를 더한 다음 모델에 입력한다!

임베딩 벡터가 Encoder의 입력으로 사용되기 전에, 다음과 같이 Positional Encoding값이 더해진다


위 그림처럼 "벡터" 단위로 더해지는게 아니라, 사실은 Embedding Vector가 모여서 만들어진 "문장 벡터 행렬"에, "Positional Encoding 행렬"이 이루어짐 

Positional Encoding의 값은 다음과 같다


pos, i, d_model?
- 

```python
def get_angles(pos, i, d_model):
    angle_rates = 1 / np.power(10000, (2 * (i//2)) / np.float32(d_model))
    return pos * angle_rates

def positional_encoding(position, d_model):
    angle_rads = get_angles(np.arange(position)[:, np.newaxis],
                          np.arange(d_model)[np.newaxis, :],
                          d_model)

    # even index : sin
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])

    # odd index : cos
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])
    pos_encoding = angle_rads[np.newaxis, ...]
    return tf.cast(pos_encoding, dtype=tf.float32)
```

### EXAMPLE
- 50 x 128의 크기를 가지는 Positional Encoding 행렬
- 입력 문장의 단어가 50개 
- 각 단어가 128차원의 임베딩 벡터를 가짐

```python
pos_encoding = positional_encoding(50, 128)
print (pos_encoding.shape) 
```

## 6. Encoder의 Multi-head Attention

### 1) Multi-head Self Attention

#### a) Self Attention?

기존의 seq2seq에서 Attention 시의 Q,K,V는
- 1) Q : t 시점의 Decoder cell에서의 Hidden state
- 2) K : 모든 시점의 Encoder cell에서의 Hidden states
- 3) V : 모든 시점의 Encoder cell에서의 Hidden states

t 시점의 Decoder state가 아니라, 전체 시점으로 일반화 하면
- 1) Q : 모든 시점의 Decoder cell에서의 Hidden states
- 2) K : 모든 시점의 Encoder cell에서의 Hidden states
- 3) V : 모든 시점의 Encoder cell에서의 Hidden states

하지만, Transformer의 Self Attention의 Q,K,V는 모두 같다!
- Q,K,V : 입력 문장의 모든 단어 벡터들

#### b) Q,K,V vector

입력받은 단어 vector로 바로 Self-Attention? NO!

각 단어 vector로부터 Q,K,V벡터를 먼저 얻어냄! ( 차원 축소가 이루어짐 )
- input 되는 단어 vector : d_model 차원 ( 논문에서 d_model = 512 )
- 차원 축소된 Q,K,V vector : d_model / num_heads 차원 ( 논문에서 num_heads = 8, 즉 64차원 )

#### c) Scaled-dot product Attention

기존과 동일! 구해낸 Q,K,V벡터로 Attention Score -> Attention Distribution -> Attention Value(Context Vector) 구함

차이점! 단순 내적 NO ( 특정 값으로 나눠준 Attention 함수를 사용함 )




아래 그림을 보면 쉽게 이해 가능하다



위 그림에서는 'I'라는 단어의 Q 벡터에 대해 연산이 이루어졌다. 

하지만, 행렬 연산으로 일괄적으로 처리하면, 이렇게 모든 Q벡터마다 일일히 따로 계산할 필요가 없다.

#### d) 행렬 연산으로 일괄 처리

[ 1. Q,K,V 행렬 구하기 ]

실제로는 벡터 간의 연산이 아니라, 행렬로 한번에 연산이 이루어지기 떄문에 Q,K,V 행렬을 우선 구한다


[ 2. 행렬 연산 통해 Attention Score 구하기 ]

Q행렬을 K행렬의 전치 행렬과 곱해준다!


[ 3. Attention Value 구하기 ]

Attention Score에 Softmax함수를 사용하고, 이에 V 행렬을 곱한다

[ 행렬의 크기 ]
입력 문장의 길이 : seq_len -> 문장 행렬의 크기는 (seq_len, d_model)....이 문장행렬을 통해 Q,K,V행렬 구함

Q & K 벡터의 크기 : d_k, V 벡터의 크기 : d_v
- Q 행렬 : (seq_len,d_k)
- K 행렬 : (seq_len,d_k)
- V 행렬 : (seq_len,d_v)

자동으로 weight 행렬의 크기도 정해진다
- Wq 행렬 : (d_model,d_k)
- Wk 행렬 : (d_model,d_k)
- Wv 행렬 : (d_model,d_v)

( 논문에서 d_k 와 d_v는, d_model / num_head로 동일 )

결과적으로, Attention Value 행렬(a)의 크기는 (seq)len, d_v)이다

#### e) Scaled-Dot Product Attention

```python
def scaed_dot(q,k,v,mask):
    matmul_qk = tf.matmul(q,k, transpose_b=True) # 1. Attention Score ( Q x V )
    
    dk = tf.cast(tf.shape(key)[-1], tf.float32) # 2. Scaling
    logits = matmul_qk / tf.math.sqrt(dk)
    
    if mask is not None:
        logits += (mask*-1e9) # 3. Masking (O,X)
    
    attention_weights= tf.nn.softmax(logits,axis=-1) # 4. Attention Distribution
    
    output = tf.matmul(attention_weights, v) # 5. Attention Score
    
    return output, attention_weights
```

### 2) Multi-head Attention

num_heads란? 왜 d_model의 차원으로 Attention을 하지 않고, 차원 축소 이후 Attention 시행?

#### "병렬 처리"

차원 축소 : d_model -> d_model / num_heads ( num_heads개 병렬 Attention을 동시에 수행! )

weight matrix (Wq,Wk,Wv)의 값이 8개 Attention Head마다 모두 다르게 됨!

#### 병렬 처리의 이유?
- 다양한 특징 capture!

병렬 Attention 모두 수행 이후, Attention head들을 concatenate! 

( 연결된 Attention Head 행렬의 크기 : (seq_len, d_model ) )

이렇게 연결된 Attention Head들에, 가중치 행렬(Wo)을 곱해서 Multi-head Attention Matrix 생성!

#### Multi-head Attention에 등장한 Weight Matrix
- 1) Wq,Wk,Wv
- 2) Wo ( Attention head 생성 후, 이에 곱하는 weight matrix )

### 3) Multi-head Attention 구현
- 1) Wq,Wk,Wv의 Dense Layer 통과
- 2) num_heads 만큼 나누는 병렬 처리
- 3) scaled dot product attention
- 4) concatenate
- 5) Wo의 Dense Layer 통과

```python
class MultiHeadAttention(tf.keras.layers.Layer):
    
    def __init__(self, d_model, num_heads, name="multi_head_attention"): # 정의하기
        super(MultiHeadAttention, self).__init__(name=name)
        self.num_heads = num_heads # 논문에서는 8
        self.d_model = d_model # 논문에서는 512
        
        assert d_model % self.num_heads == 0
        
        self.depth = d_model // self.num_heads
        
        self.query_dense = tf.keras.layers.Dense(units=d_model) #WQ
        self.key_dense = tf.keras.layers.Dense(units=d_model) #WK
        self.value_dense = tf.keras.layers.Dense(units=d_model) #WV

        self.dense = tf.keras.layers.Dense(units=d_model) #WO

    def split_heads(self, inputs, batch_size): # 아래의 call 함수에서 헤드를 나누기 위해서 호출
        inputs = tf.reshape(inputs, shape=(batch_size, -1, self.num_heads, self.depth))
        return tf.transpose(inputs, perm=[0, 2, 1, 3])

    def call(self, inputs):
        query, key, value, mask = inputs['query'], inputs['key'], inputs['value'], inputs['mask']
        batch_size = tf.shape(query)[0]

        # 1) Wq,Wk,Wv의 Dense Layer 통과
        query = self.query_dense(query) # (batch_size, seq_len, d_model) 
        key = self.key_dense(key) # (batch_size, seq_len, d_model)
        value = self.value_dense(value) # (batch_size, seq_len, d_model)

        # 2) num_heads 만큼 나누는 병렬 처리
        query = self.split_heads(query, batch_size) # (batch_size, num_heads, seq_len, d_model/num_heads)
        key = self.split_heads(key, batch_size) # (batch_size, num_heads, seq_len, d_model/num_heads)
        value = self.split_heads(value, batch_size) # (batch_size, num_heads, seq_len, d_model/num_heads)

        # 3) Scaled dot product attention
        scaled_attention = scaled_dot_product_attention(query, key, value, mask)
        scaled_attention = tf.transpose(scaled_attention, perm=[0, 2, 1, 3])

        # 4) Concatenate
        concat_attention = tf.reshape(scaled_attention,
                                      (batch_size, -1, self.d_model))

        # 5_ Wo의 Dense Layer 통과
        outputs = self.dense(concat_attention)

        return outputs 
```
