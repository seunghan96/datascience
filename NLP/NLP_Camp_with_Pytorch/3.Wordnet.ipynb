- one hot vector로는 그런 단어의 특징 발견 hard!

### hypernyms(상위어) 찾기

```python
wn.synsets('people')
```

```python
wn.synsets('people')[0]
```

```python
def hypernyms(word) :
    current_node = wn.synsets(word)[0]
    yield current_node
    
    while True:
        try:
            current_node = current_node.hypernyms()[0]
            yield current_node
        except IndexError:
            break
```

### ex) 각 단어에 대한 상위어는?

```python
[h for h in hypernyms('people')]
```

```python
[h for h in hypernyms('policeman')]
```

```python
[h for h in hypernyms('chocolate')]
```
